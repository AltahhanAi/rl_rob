```python
from rl.linear import *
from rl.neural import *  # DQN, DDQN, nnSarsa, nnQlearn, demoGame
```

```python
def imaze(**kw):
    return maze(iGrid, **kw)

env = imaze(reward='sparse')
dqn = DQN(env=env, ε=.1, γ=.98, seed=1, episodes=30, q0=0,
          α=1e-4, create_Wn=True,
          trunk=[(8, 4, 2), (4, 4, 4)], nF=env.nS,
          nbuffer=1000, nbatch=64, t_Qn=100,
          clipModel=True, **demoGame).interact()
```

Swap `DQN` for `DDQN` to fix value overestimation, or `nnSarsa`/`nnQlearn` for the online variants.
