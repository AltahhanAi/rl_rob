```python
import gymnasium as gym
```

```python
# Create the HalfCheetah environment
env = gym.make("HalfCheetah-v5", render_mode=None)  # or "human" for a viewer window

# Reset the environment
obs, info = env.reset()
print("Initial observation shape:", obs.shape)
```

```python
for step in range(10):  # minimal interaction: 10 steps
    # Sample a random action from the action space
    action = env.action_space.sample()

    # Take a step in the environment
    obs, reward, terminated, truncated, info = env.step(action)

    print(f"Step {step}: reward={reward}, terminated={terminated}, truncated={truncated}")

    if terminated or truncated:
        # If episode is done, reset
        obs, info = env.reset()
        print("Environment reset")

env.close()
```
